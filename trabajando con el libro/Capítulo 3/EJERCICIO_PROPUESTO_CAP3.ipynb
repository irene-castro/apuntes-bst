```
# Leer el csv del capitulo 2 y obtener la estructura del schema dado por defecto
from pyspark.sql.types import *

cap2_file="/FileStore/shared_uploads/irene.castro@bosonit.com/mnm_dataset.csv"
cc2v=spark.read.csv(cap2_file,header=True)

print(cc2v.printSchema())
```

```
# Utilizando el mismo ejemplo del capítulo 2 para guardar en parquet, y guardar los archivos en formato:
# i) JSON
# ii) CSV (dándole otro nombre para no reescribir el fichero de origen)
# iii) AVRO

json_path="/FileStore/shared_uploads/irene.castro@bosonit.com/jsonpruebaa"
cc2v.write.format("json").save(json_path)
```

```
csv_path="/FileStore/shared_uploads/irene.castro@bosonit.com/csvprueba"
cc2v.write.format("csv").save(csv_path)
```

```
avro_path="/FileStore/shared_uploads/irene.castro@bosonit.com/avroprueba"
cc2v.write.format("avro").save(avro_path)
```
